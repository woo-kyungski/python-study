#### *숫자형 연산 Operation
- 산술 연산

````python
print(10 + 3) # 덧셈
print(10 - 3) # 뺄셈
print(10 * 3) # 곱셈
print(10 / 3) # 나눗셈
````

- 복합 연산

````python
c = 10
print(c)
````

````python
c += 3 # += : c = c + 3
print(c)

c -= 3 # -= : c = c - 3
print(c)

c *= 3 # *= : c = c * 3
print(c)

c /= 3 # /= : c = c / 3
print(c)
````

#### 2) 불리언 Boolean

- 참 True / 거짓 False

#### *불리언 연산과 비교연산

- 비교연산 Comparisons
    - < : 작다
    - <= : 작거나 같다
    - \> : 크다
    - \>= : 크거나 같다
    - == : 같다
    - != : 다르다

- 논리 연산 Logical Operations 
    - and : 하나라도 거짓인 값이 존재할 때, 결과는 False
    - or :  둘다 모두 거짓인 값일 때, 결과는 False
    - not : 기존의 값을 반대로 바꿀 때, 결과는 기존 값 반대 (True -> False, False -> True)

````python
t = True
f = False
````

````python
# and
print (t and t) # 
print (t and f) # 
print (f and f) # 

# or
print(t or t) # 
print(t or f) # 
print(f or f) # 

# not
print(not t) # 
print(not f) # 
````
