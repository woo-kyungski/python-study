#### 1) 함수 생성
```python    
def 함수명([매개변수 parameter, ...]):
    수행코드
    ...
    [return 반환값]
```


````python
# 입력값이 없고, 반환값이 없는 함수(입력값X, 반환값X)

def hello1():
	print('Hello, Phtyon') # Hello, Python
````

````python
hello1()
````

````python
# 입력값이 없고, 반환값이 있는 함수(입력값X, 반환값O)

def hello2():
	return 'Hello, Python' # Hello, Python
                        #'return'값을 가지고 있다가 전달해 줌
````

````python
print(hello2())
````

````python
# 입력값이 있고, 반환값이 없는 함수(입력값O, 반환값X)

language = 'Python'

def hello3(language):
	print('Hello, ', language) # Hello, Python
````

````python
hello3('얼럴럴')
````

````python
# 입력값이 있고, 반환값이 있는 함수(입력값O, 반환값O)

language = 'Python'

def hello4(language):
	return 'Hello, ' + language # Hello, Python
````

````python
print(hello4('wksk')) #따옴표가 있음 --> print 사용해서 결과를 제대로 확인해야 함
````

#### 3) 활용
- 다중 결과값

````python
# +
def plus(a, b) : 
  return a +b
````

````python
print(plus(1,2))
````

````python
# +, -
def plus_minus(a,b) :
  return a + b 
  return a - b   # (실행 안됨) / return = '여기에서 실행이 종료된다' / 한 개만 사용하기
````

````python
print(plus_minus(5,3))
````

````python
# +, -
def plus_minus2(a,b) :
  return a + b, a - b  
````

````python
print(plus_minus2(5,3))
````

````python
# +, -, * / 기본 사칙연산에 대한 결과값을 반환하는 함수 basic_cal
def basic_cal(a,b) :
  return a + b, a - b, a * b, a / b  
````

````python
# basic_cal_op
def basic_cal_op(a, b, op):
  if op == '+':
    return a + b
  elif op == '-':
    return a - b
  elif op == '*':
    return a * b
  elif op == '/':
    return a / b
````

````python
print(basic_cal_op(5, 3, '*'))
````

- 다중 입력값
    - *args : 다중 매개변수
    - **kwargs : 키워드 매개변수

````python
# 다중 매개변수 : 매개변수가 여러개
def print_data(a, b, c) : 
  print(a)
  print(b)
  print(c)
````

````python
print_data(1,2,3)
````

````python
def print_data2(*args) : # '*'가 포인트임: '무한하게 데이터가 들어갈 수 있다'
  print(args)
````

````python
print_data2(1,2,3,4,5,6,7,8)
````

````python
# 한줄씩 결과값이 출력되도록 하려면?
def print_data3(*args):
  for arg in args :
    print(arg)
````

````python
print_data3(1,2,3,4,5,6,7)
````

- *결과 <br>

````python
# 키워드 매개변수 : 키와 값 한 쌍으로 구성 되어 있는 데이터 형식
def print_name_age(**kwargs): # keyword arguments
  print('이름 : ', kwargs['name'])
  print('나이 : ', kwargs['age'])
````

````python
print_name_age(name = 'playdata', age = '6')
````

나이 :  6
