## 숫자만을 원해
- 숫자를 다루어야 하는 일이 생겼습니다. 그런데 데이터를 보니 콤마가 포함이 되어있네요... 나는 숫자만 필요한데... 숫자만 반환해주는 함수를 만들어 주세요.
    - remove_comma(data)

```python
data = '23,524,414'
```

```python
# 함수 작성
def remove_comma(data):
  return int(data.replace(',',''))
```

```python
remove_comma('23,524,414')
```

```python
# 2-2
def remove_comma2(data):
  data_new = data.split(',')
  num = ''
  for i in data_new:
    num += i
  return int(num)
```

### 결과
23524414

## 나의 평균 푸쉬업 개수는?
- 건강을 위해 20일동안 푸쉬업을 했습니다. 20일동안 나의 평균 푸쉬업 개수는 몇개일까요?
    - avg_push(nums)

```python
nums = [6, 2, 5, 7, 9, 6, 3, 8, 11, 15, 9, 16, 16, 12, 18, 20, 25, 27, 23, 31]
```

```python
# 함수 작성

def avg_push(nums):
  total_sum = 0
  for num in nums:
    total_sum = total_sum + num

  return total_sum/len(nums)
 
```

```python
print(avg_push(nums))
```

### 결과
13.45

```python
# 함수 작성
no_snacks = {'미니비스킷': [2580, 1575], '별사탕건빵': [980, 1025], '바삭한콘칩': [1950, 503], '담백한통감자': [1680, 524], '고르곤졸라소프트콘': [1280, 690]}
total_price = 0
total_kcal = 0
def cal_total_price_kcal(no_snacks, total_price, total_kcal):
  for i in no_snacks:
    total_price += no_snacks[i][0]
    total_kcal += no_snacks[i][1]
  print("총가격 : ", total_price, "원")
  print("총열량 : ", total_kcal, "kcal")
  
```

```python
cal_total_price_kcal(no_snacks, total_price, total_kcal)
```
