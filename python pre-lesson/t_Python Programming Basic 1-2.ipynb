#### 2) 리스트 내부 데이터 추출 : 인덱싱, 슬라이싱

````python
a = [1,2,3,4,5]
print(a)
````

````python
# 리스트 길이 구하기 : len(리스트 변수)
print(len(a))
````

````python
# 3만 추출하고 싶다면?
print(a[2])
````

````python
# 연산도 가능?
a[1] +a[3]
````

````python
# 3부터 5까지추출 
a[2:]
````

````python
b = [1, 2, 3, ['a', 'b', 'c', ['one', 'two', 'three']]]
````

````python
# 'b' 문자만 출력하기
print(b[3][1])
````

````python
# 'three' 문자열 출력하기
print(b[3][3][2])
````

#### 3) 리스트 수정 및 삭제

````python
programming_lang = ['python', 'java', 'javascript']
print(programming_lang)
````

- 요소 수정
    - 해당 인덱스에 존재하는 요소 값을 직접 변경

````python
# python -> pyhton3으로 변경하려면?
programming_lang[0] = 'python3'
programming_lang
````

- 요소 삭제
    - del 리스트내 요소 : 리스트 내부 지정 요소를 삭제

````python
# 삭제 전
print(programming_lang)
````

````python
# 삭제 후
# 삭제 test1
del programming_lang[2]
programming_lang
````

````python
programming_lang = ['python', 'java', 'javascript']
print(programming_lang)
````

````python
# 삭제 test2 : 슬라이싱으로 요소 삭제
print(programming_lang)
print('삭제 전 : ', programming_lang)

del programming_lang[1:]
print('삭제 후 : ', programming_lang)
````

#### 4) 리스트 활용

- 요소 추가 
    - append(data) : 리스트 맨 뒤에, 지정한 요소를 추가
    - insert(idx, data) : 리스트 idx 위치에 지정한 요소를 추가

````python
fruits = ['apple', 'banana', 'grape']
print(fruits)
````

````python
# append
fruits.append('kiwi')
fruits
````

````python
# insert
fruits.insert(2, "orange")
````

````python
fruits
````

- 요소 삭제
    - remove(data) : 리스트 내 지정한 요소와 같은 첫번째 내용을 삭제 

````python
fruits = ['apple', 'banana', 'grape']
print(fruits)
````

````python
fruits.append('banana')
fruits
````

````python
# remove: 앞에 있는 것부터 삭제 --> 데이터가 없다면 오류 발생
fruits.remove('banana')
fruits
````

- 요소 추출 및 삭제
    - pop([idx])
    ```
        * 인덱스 지정시, 리스트 내 지정 인덱스 요소 반환 후 리스트에서 삭제
        * 인덱스 미지정시, 리스트 내 맨 뒤 요소 반환 후 리스트에서 삭제
    ```

````python
fruits = ['apple', 'banana', 'grape']
print(fruits)
````

````python
fruits.pop()
````

````python
print(fruits)
````

````python
fruits.pop(0)
````

````python
print(fruits)
````
