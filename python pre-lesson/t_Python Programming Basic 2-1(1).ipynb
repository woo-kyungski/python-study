## 집합 자료형 Set Types
### 1.집합 Set

- 집합을 처리하기 위해 사용되는 자료형
- 중복이 허용되지 않고, 순서가 존재하지 않음

#### 1) 집합 생성
- 방법
    - 튜플 변수 = ([데이터1], [데이터2], ...)

```python
set1 = set()
```

```python
set2 = set('abc')
```

```python
set3 = set([1,2,3])
```

```python
set4 = set('programming')
```

```python
print(set4)
```

#### 2) 집합 데이터 추가
- add() : 집합에 하나의 데이터(숫자, 문자열, 튜플)만 추가
- update(*others) : 집합에 여러개의 값을 추가

```python
set5 = {1, 2, 3, 4, 5}
print(set5, type(set5))
```

```python
# add
set5.add('ab') #add로 추가될 시 한 개의 데이터로 추가됨(나눠지지 않음)
print(set5)
```

```python
# update
set5.update('xyz') #여러 데이터를 한 번에 추가 가능
print(set5)
```

#### 3) 집합 데이터 삭제
- remove(elem) : 집합에서 지정요소 삭제, 요소 미존재시 에러 발생
- discard(elem) : 집합에서 지정요소 삭제,  요소 미존재시 에러 발생하지 않음

```python
set6 = {1, 2, 3, 4, 5}
print(set6)
```

```python
# remove
set6.remove(3)
print(set6)
```

```python
# discard
set6.discard(4) #해당 삭제할 데이터가 없는 상태에서 실행 시켜도 에러가 발생하지 않음
print(set6)
```

#### 4) 집합 활용

- 4-1) 합집합(union)
- 4-2) 교집합(intersection)
- 4-3) 차집합(difference) 

```python
market1 = {'apple', 'banana', 'grape', 'orange', 'kiwi'}
market2 = {'apple', 'pineapple', 'orange', 'strawberry'}
market3 = {'orange', 'strawberry', 'cherry'}

print('market1 : ', market1)
print('market2 : ', market2)
print('market3 : ', market3)
```

```python
# 합집합
print(market1.union(market2))
print(market1|market2)
```

```python
# 교집합
print(market1.intersection(market3))
print(market1 & market3)
```

```python
# 차집합
# market1에서만 존재
print(market1.difference(market2))
print(market1-market2)

# market2에서만 존재
print(market2.difference(market1))
print(market2-market1)
```

```python
# market1, 2, 3에 모두 존재하는 과일은?
print(market1 & market2 & market3)
print(market1.intersection(market2.intersection(market3)))
```

## 매핑 자료형 Mapping Types
### 1.딕셔너리 Dictionary

- 키 Key와 값 Value를 한 쌍으로 갖는 자료형
- 키를 통해 값을 반환 받을 수 있음

#### 1) 딕셔너리 생성
- 방법
    - 딕셔너리 변수 = {[Key1 : Value1, Key2 : Value2, ...]}

```python
dict1 = dict([('one','1'),('two','2')])
dict1
```

```python
dict2 = {'one':1, 'two':2}
dict2
```

#### 2) 딕셔너리 내부 값 추출
- 딕셔너리 변수이름[key]
- 딕셔너리.get[key]

```python
# 'two'값 추출하기
print(dict2['two'])
```

```python
# get
print(dict2.get('two'))   # 잘못된 키 값 입력 시 오류 대신 None값 출력
```

#### 3) 딕셔너리 수정 및 삭제
- 수정
    - 딕셔너리이름[키] = new 데이터

```python
dict4 = {'apple' : 3, 'banana' : 9, 'orange' : 1}
print(dict4)
```

```python
# orange 추가 구입
dict4['ornage'] += 2
print(dict4)                 #딕셔너리는 값을 수정할 수 있다 
```

- 삭제
    - del 딕셔너리이름[키]

```python
print(dict4)
```

```python
# apple 구매하지 않음 1) apple : 0
                  #   2) 딕셔너리에서 삭제 Del
del dict4['apple']    # 삭제 대상이 없을 시 에러 메시지
dict4 
```

- 추가
    - 딕셔너리이름[새로운 키] = 데이터

```python
print(dict4)
```

```python
# plum 1개 추가 구매
dict4['plum'] = 1
dict4
```

#### 4) 주의사항 및 활용
- 키 Key는 '고유'한 값

```python
dict5 = {'a' : 1, 'b' : 2, 'c' : 3}
print(dict5, type(dict5))
```

```python
# 수정 불가능
dict5['a'] = 4
dict5
```

- 키 Key 혹은 값 Value 모두 반환하기
    - keys() : 딕셔녀리의 키만 dict_keys 객체로 반환
    - values() : 딕셔녀리의 값만 dict_values 객체로 반환
    - items() : 딕셔너리의 키, 값 쌍으로 dict_items 객체로 반환

```python
dict5 = {'a' : 1, 'b' : 2, 'c' : 3}
print(dict5, type(dict5))
```

```python
# keys
dict5.keys()
```

```python
# values
dict5.values()
```

```python
# items
dict5.items()
```

```python
for item in dict5.items() :
  print(item[0])
```

```python
# 키 존재여부 확인
dict5 = {'a' : 1, 'b' : 2, 'c' : 3}
print(dict5, type(dict5))
```

```python
# in
if 'c' in dict5 :
  print(dict5['c'])
```
