## 핫딜 : 네이버 쇼핑 
- https://shopping.naver.com/hotdeal/p/index.naver

````python
import requests
from bs4 import BeautifulSoup

response = requests.request('GET', 'https://shopping.naver.com/hotdeal/p/index.naver')
soup = BeautifulSoup(response.content, 'html.parser')

hot_keywords = soup.find('ol', id ='best100_keyword1')
for keyword in hot_keywords.find_all('li') :
  full_word = ''
  for word in keyword.get_text().split()[1:-1] :
    full_word = full_word + ' ' + word
  print(full_word)
````

````python
import requests
from bs4 import BeautifulSoup

response = requests.request('GET', 'https://www.forbes.com/advisor/travel-rewards/top-50-best-places-to-visit/')
soup = BeautifulSoup(response.content, 'html.parser')
````

````python
top_places_list = []
top_places = soup.find_all('h2') #wksk

for places in top_places :
  top_places_list.append(places.get_text()) 
  
  print(top_places_list)


````

````python
import requests
from bs4 import BeautifulSoup

def crawling (url, name):
  response = requests.request('GET', url)
  soup = BeautifulSoup(response.content, 'html.parser')

  hot_places = soup.select(name)
  list = []
  for keyword in hot_places:
    list.append(keyword.get_text())

  return list

url = 'https://www.forbes.com/advisor/travel-rewards/top-50-best-places-to-visit/'
name = 'h2'

print(crawling(url, name))
````

````python
import requests as rq
from bs4 import BeautifulSoup

url = 'https://www.forbes.com/advisor/travel-rewards/top-50-best-places-to-visit/'
def dataCrawling(url):
  rps = rq.request('GET',url)
  soup = BeautifulSoup(rps.content, 'html.parser')
  visits = soup.find_all('h2')
  visit_final=[]
  for visit in visits:
    visit_final.append(visit.get_text())
  return visit_final

print(dataCrawling(url))
````
