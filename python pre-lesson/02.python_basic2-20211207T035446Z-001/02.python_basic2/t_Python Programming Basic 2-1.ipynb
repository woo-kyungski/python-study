## 집합 자료형 Set Types
### 1.집합 Set

- 집합을 처리하기 위해 사용되는 자료형
- 중복이 허용되지 않고, 순서가 존재하지 않음

#### 1) 집합 생성
- 방법
    - 튜플 변수 = ([데이터1], [데이터2], ...)

#### 2) 집합 데이터 추가
- add() : 집합에 하나의 데이터(숫자, 문자열, 튜플)만 추가
- update(*others) : 집합에 여러개의 값을 추가

```python
set5 = {1, 2, 3, 4, 5}
print(set5, type(set5))
```

```python
# add
```

```python
# update
```

#### 3) 집합 데이터 삭제
- remove(elem) : 집합에서 지정요소 삭제, 요소 미존재시 에러 발생
- discard(elem) : 집합에서 지정요소 삭제,  요소 미존재시 에러 발생하지 않음

```python
set6 = {1, 2, 3, 4, 5}
print(set6)
```

#### 4) 집합 활용

- 4-1) 합집합(union)
- 4-2) 교집합(intersection)
- 4-3) 차집합(difference) 

```python
market1 = {'apple', 'banana', 'grape', 'orange', 'kiwi'}
market2 = {'apple', 'pineapple', 'orange', 'strawberry'}
market3 = {'orange', 'strawberry', 'cherry'}
print('market1 : ', market1)
print('market2 : ', market2)
print('market3 : ', market3)
```
