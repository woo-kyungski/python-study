## 클래스 Class 와 객체 Object

### 1.클래스 Class 

- 객체를 만들기 위한 설계도

- 사용 이유
    - 기능과 데이터를 한 번에 묶어 관리함으로써 용이하게 관리

- 구조 및 문법
```python
class 클래스이름:
    
    [변수이름 = 데이터]
    
    [def 메서드이름(self, [매개변수, ...]):
         실행구문]
    ...
```
    - 클래스는 대문자로 시작한다.
    - 클래스 : 속성 Attribute + 메서드 Method
        - 속성(Attribute)란? 클래스 내부에 포함되는 변수
        - 메서드(Method)란? 클래스 내부에서 구현된 함수
            * 주의점 : 메서드의 첫번째 매개변수로 self를 명시한다.

### 2.객체 Object 

- 물리적으로 실제 존재하는 물건 혹은 추상적인 개념
- 객체 지향 프로그래밍(OOP : Object-Oriented Programming) : 객체를 활용하는 프로그래밍 기법
- 문법
```python
    # 객체 생성
    객체 변수명 = 클래스명()

    # 속성 또는 메서드 호출
    객체 변수명.속성 또는 객체 변수명.메서드
```

### *자동차 Car 클래스 만들기
- 속성 : 브랜드이름, 모델이름, 최대속도, 연비, 가격, ...
- 메서드 : 빨라진다, 느려진다, 멈추다, 우측으로 가다, ...

````python
# 클래스
# 자동차 클래스 생성
class Car:
  # attribute
  model_name = 'G'
  color = 'red'
  
  # method
  def start_engine(self):
    print('시동이 걸렸습니다.')

  def stop_engine(self):
    print('시동이 꺼졌습니다.')
````

````python
# 객체(인스턴스) 생성
my_car = Car()
````

````python
my_car.color
````

````python
my_car.start_engine()
````

````python
my_car2 = Car()
````

````python
my_car2.color
my_car2.color = 'blue'
my_car2.color
````

- 스페셜 메서드 Special Method or 매직 메서드 Magic Method
    - 메서드이름 앞 뒤로 __(밑줄 두개)가 붙어, 자동으로 호출해주는 메서드
    - \__init__ 메서드
        - 인스턴스(객체) 생성시 자동 호출
        - 인스턴스 초기화

````python
# 자동차 클래스 Car
# 클래스
# 자동차 클래스 생성
class Car:
  def __init__(self, model_name = '', color = ''):
  	# attribute
  	self.model_name = model_name
  	self.color = color
  
  # method
  def start_engine(self):
    print('시동이 걸렸습니다.')
  
  def stop_engine(self):
    print('시동이 꺼졌습니다.')
````

````python
my_car3 = Car('wksk','yellow')
````

````python
my_car3.color
````

````python
my_car3.model_name
````
