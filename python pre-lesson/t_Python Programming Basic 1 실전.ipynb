## 담임 선생님의 채점을 도와주세요, 플리즈
- 다음은 A반 학생들의 중간고사 시험 성적입니다. 성적 평균을 구해주세요. (소수점 2째자리까지만 출력해주세요.)
    - scores = [53, 67, 42, 99, 29, 87, 44, 75, 23, 76, 49, 63, 92, 79, 70]

### 결과

```python
total_score = 0

for score in scores :
  total_score = total_score + score

print('%.2f' % (total_score / len(scores)))
```

## 종일 파일 파일정리만 할겐가
- 파일 목록에서 파일 확장자가 xlsx인 파일들만 출력해주세요
    - files = ['movie.xls', 'report.doc', 'stock.xlsx', 'print.png', 'database.hwp', 'voca.xlsx']

```python
files = ['movie.xls', 'report.doc', 'stock.xlsx', 'print.png', 'database.hwp', 'voca.xlsx']
```

### 결과
stock.xlsx <br>
voca.xlsx

```python
for i in files : #wksk
  if i[-5:] == '.xlsx' :
   print(i)
```

```python
for file in files :
  if file.split('.')[1] =='xlsx' : #스플릿 메서드 활용
    print(file)
```
